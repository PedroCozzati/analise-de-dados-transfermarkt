# Filtro para selecionar apenas jogadores que estavam ativos a partir em 2023 e que jogam na Bundesliga

```python
current_player_data = player_data[player_data['last_season']>=2023]
```

```python
current_player_data_bundesliga = current_player_data[current_player_data['current_club_domestic_competition_id']=='L1']
```

# Preencher valores nulos para o "pé dominante" do jogador com o próximo valor válido (ffill)

```python
current_player_data_bundesliga.foot.value_counts(normalize=True)
```

```python
current_player_data_bundesliga['foot'].fillna(method='ffill',inplace=True)
```

```python
current_player_data_bundesliga.foot.value_counts(normalize=True)
```

# Preencher valores nulos para o "agente" do jogador

```python
current_player_data_bundesliga['agent_name'].fillna('Player without an agent',inplace=True)
```

```python
agent_dict = current_player_data_bundesliga['agent_name'].value_counts()
```

```python
agent_dict[:10]
```

```python
fig, ax = plt.subplots()

labels = agent_dict.keys()[:10]
counts = agent_dict[:10]

bar_colors = ['red', 'blue', 'green','purple', 'blue', 'green','purple', 'blue', 'green','purple']

ax.barh(labels, counts, color=bar_colors,align='edge')

ax.set_ylabel('Agentes')
ax.set_title('Top 10 Agentes de jogadores na Bundesliga')
ax.invert_yaxis()
plt.show()
```
