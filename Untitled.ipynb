# Filtro para selecionar apenas jogadores que estavam ativos a partir em 2023 e que jogam na Bundesliga

```python
current_player_data = player_data[player_data['last_season']>=2023]
```

```python
current_player_data_bundesliga = current_player_data[current_player_data['current_club_domestic_competition_id']=='L1']
```

# Tratamento de dados nulos

## Preencher valores nulos para o "pé dominante" do jogador com o ultimo valor válido (ffill)

```python
current_player_data_bundesliga.foot.value_counts(normalize=True)
```

```python
current_player_data_bundesliga['foot'].fillna(method='ffill',inplace=True)
```

```python
current_player_data_bundesliga.foot.value_counts(normalize=True)
```

```python
current_player_data_bundesliga.isnull().sum()
```

## Preencher valores nulos para o "agente" do jogador

```python
current_player_data_bundesliga['agent_name'].fillna('Player without an agent',inplace=True)
```

```python
current_player_data_bundesliga['has_an_agent'] = current_player_data_bundesliga['agent_name']!='Player without an agent'
```

```python
current_player_data_bundesliga
```

```python
agent_dict = current_player_data_bundesliga[current_player_data_bundesliga['has_an_agent']==True]
```

```python
agent_list = agent_dict['agent_name'].value_counts()
```

```python
agent_list[:10]
```

```python
has_an_agent = current_player_data_bundesliga['has_an_agent'].value_counts()
```

```python
has_an_agent
```

```python
fig, ax = plt.subplots(2,1)

labels = agent_list.keys()[:10]
counts = agent_list[:10]

pie_counts = current_player_data_bundesliga['has_an_agent'].value_counts()

bar_colors = ['red', 'blue', 'green','purple', 'blue', 'green','purple', 'blue', 'green','purple']

ax[0].barh(labels, counts, color=bar_colors)

ax[0].set_ylabel('Agentes')
ax[0].set_title('Top 10 Agentes de jogadores na Bundesliga (2023)')
ax[0].invert_yaxis()

ax[1].pie(pie_counts,labels=['Possui Agente','Não Possui'],autopct='%1.1f%%')

# ax[1].set_ylabel('Agentes')
ax[1].set_title('Porcentagem de jogadores com agentes (2023)')
ax[1].invert_yaxis()

fig.tight_layout()

plt.show()
```
